```python

import pandas as pd
```

```python
# Import "datos-RS" and convert it into a new DATAFRAME
rs = pd.read_excel(r'C:\Users\Alvaro\OneDrive\Documentos\Primer proyecto\codigo\Rinconcito_del_sabor\datos-RS.xlsx')
rs.head()
```

```python
# View the information of the DataFrame
rs.info()
```

```python
# Rename the column "Tipo de Ingreso/Gasto" to "Tipo"
rs.rename(columns={'Tipo de Ingreso/Gasto': 'Tipo'}, inplace=True)

# Display the DataFrame with the new column name
rs
```

```python
# Rename the column "Tipo Op" to "Operacion"
rs.rename(columns={'Tipo Op': 'Operacion'}, inplace=True)

# Display the DataFrame with the new column name
rs
```

```python
# Filter the rows that have null values in the "Salida" column
nulls = rs[rs['Salida'].isnull()]

# Display the complete rows that have null values in the "Salida" column
print(nulls)
```

```python
# Calculate the average of values in the "Salida" column where the "Concepto" column contains "gas"
promedio_gas = rs[rs['Tipo'].str.contains('gas', case=True)]['Salida'].mean()

# Fill the null values in the "Salida" column with the calculated average
rs['Salida'].fillna(promedio_gas, inplace=True)

# Display the DataFrame
rs  
```

```python
# Display row 1511
fila_1511 = rs.iloc[1509]  

print(fila_1511)
```

```python
# Obtain the days of the week as numbers (0: Monday, 1: Tuesday, ...)
rs['DiaSemanaNum'] = rs['Fecha'].dt.dayofweek

# Map the numbers to day names of the week
dias_semana = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
rs['DiaSemana'] = rs['DiaSemanaNum'].map(lambda x: dias_semana[x])

# Display the DataFrame with the new column
print(rs)
```
